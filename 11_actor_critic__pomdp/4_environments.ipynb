OpenAI's Gym
------

A toolkit for developing and comparing reinforcement learning algorithms.

Environments should only be implemented once. 

```python
import gym
```

OpenAI's Environments
--------

https://gym.openai.com/envs/#classic_control 

https://github.com/openai/gym/blob/master/docs/environments.md#third-party-environments

```python
env = gym.make('Acrobot-v1') # Swing up a two-link robot
```

```python
env.reset()

for _ in range(1000):
    env.render()
    env.step(env.action_space.sample()) # take a random action

env.close()
```

OpenAI's Observations
------

https://gym.openai.com/docs/#observations 

The environment is like an API that returns values at a state

```python
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```
